# sklearn imputer is also a method for filling missing values. Testing follows.

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'median')
```

```python
df_to_impute = df.select_dtypes(['int','float'])
```

```python
df_to_impute
```

```python
X = df_to_impute.dropna(subset = ['Total Cost']).drop('Total Cost', axis = 1)
```

```python
y = df_to_impute.dropna(subset = ['Total Cost'])['Total Cost']
```

```python
imputer.fit(X)
```

```python
X = imputer.transform(X)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X).isna().sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectKBest
```

```python
pipe = Pipeline([('imputer',SimpleImputer()),
                 ('kbest',SelectKBest()),
                 ('lgr',LinearRegression())])
```

```python
pipe.fit(X,y)
```

# Python library SMOTE can be also used to populate samples

```python
df['FSV CMSI Flag'].value_counts(normalize = True)
```

```python
yes = df.loc[df['FSV CMSI Flag'] == 'Y']
```

```python
no = df.loc[df['FSV CMSI Flag'] == 'N']
```

```python
no_sample = no.sample(951)
```

```python
combined = pd.concat( [yes, no_sample]).dropna(subset = ['Total Cost'])
```

```python
X = combined[['Total Cost']]
y = combined[['FSV CMSI Flag']]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(X,y)
```

```python
lgr.predict_proba(X)
```

```python
lgr.score(X,y)
```
