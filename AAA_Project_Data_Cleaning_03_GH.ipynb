# Using Linear Regression to test the correlation between features 'City','Motorcycle Indicator','FSV Credit Card Flag' to 'Total Cost'

```python
X = df[['City','Motorcycle Indicator','FSV Credit Card Flag','Total Cost']].dropna()
```

```python
y= X['Total Cost']
```

```python
#X.drop('Total Cost', axis =1)
X = X.drop('Total Cost', axis =1)
```

```python
X.head()
```

# Different transforms can be used together using Pipeline of sklearn.

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.pipeline import Pipeline
```

```python
ohe= OneHotEncoder()
```

```python
X_numbers = ohe.fit_transform(X)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_numbers,y)
```

```python
pipe = Pipeline([('ohe', OneHotEncoder(handle_unknown='ignore')), ('lr', LinearRegression())])
```

```python
pipe.fit(X,y)
```

# Check the score of the Pipe which involves two transformers, ohe and lr

```python
pipe.score(X,y)
```

```python
pipe.named_steps
```

```python
ohe.get_feature_names()
```
